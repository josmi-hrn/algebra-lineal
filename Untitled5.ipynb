```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from cv2 import imdecode, IMREAD_GRAYSCALE
from google.colab import files

# 1. INTERFAZ DE CARGA DE IMAGEN INTERACTIVA
print("====================================================")
print("  PROCESAMIENTO CON ALTO BRILLO Y CONTRASTE          ")
print("====================================================\n")

uploaded = files.upload()

if len(uploaded) == 0:
    print("[ERROR]: No se seleccionó ningún archivo.")
else:
    filename = list(uploaded.keys())[0]
    file_bytes = np.frombuffer(uploaded[filename], np.uint8)
    img_original = imdecode(file_bytes, IMREAD_GRAYSCALE)

    print(f"\n[ÉXITO]: Imagen '{filename}' cargada correctamente.")

    # --- FLUJO DE CALIBRACIÓN FUERTE ---

    # Paso 1: Más Brillo (+20) para iluminar la placa de forma general
    img_b = np.clip(img_original.astype(float) - 30, 0, 255)

    # Paso 2: Más Contraste (*1.40) para marcar fuertemente los blancos sobre los negros
    img_c = np.clip(np.round(img_b * 1.85), 0, 255)

    # Paso 3: Extracción de contornos con Kernel Laplaciano
    kernel_laplaciano = np.array([
        [ 0, -1,  0],
        [-1,  4, -1],
        [ 0, -1,  0]
    ])
    bordes_puros = ndimage.convolve(img_c, kernel_laplaciano, mode='constant', cval=0)

    # Paso 4: Fusión de Nitidez (Acoplamos el 25% de los bordes calculados)
    img_transformada = np.clip(img_c + 0.25 * bordes_puros, 0, 255).astype(np.uint8)

    # 3. CÁLCULO DE LA MATRIZ DE DIFERENCIA NETA
    dif_neta = img_transformada.astype(int) - img_original.astype(int)

    # 4. ENTORNO VISUAL COMPARATIVO 2D
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    plt.subplots_adjust(wspace=0.25)

    # Imagen Original (Antes)
    axes[0].imshow(img_original, cmap='gray', vmin=0, vmax=255)
    axes[0].set_title("A. Imagen Original", fontsize=12)
    axes[0].axis('on')

    # Imagen Procesada (Después)
    axes[1].imshow(img_transformada, cmap='gray', vmin=0, vmax=255)
    axes[1].set_title("B. Imagen con Brillo(c=-30), contraste(k=1.85) y enfoque(constante o elemento medio=4)", fontsize=12)
    axes[1].axis('on')

    # Matriz de Variación Numérica (Diferencia)
    im_dif = axes[2].imshow(dif_neta, cmap='jet')
    axes[2].set_title("C. Matriz de Diferencia Neta", fontsize=12)
    axes[2].axis('on')
    fig.colorbar(im_dif, ax=axes[2], label='Delta de Intensidad Acumulado')

    plt.suptitle("Punto 5: Transformación y mejoramiento de imagen completa", fontsize=14, y=0.98)
    plt.show()

    # Guardar archivo de evidencia para el reporte
    fig.savefig('evidencia_torax_alto_contraste.png', bbox_inches='tight')
    print("\n[INFO]: El panel comparativo se exportó como 'evidencia_torax_alto_contraste.png'")
```
