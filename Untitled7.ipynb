```python
import io
import numpy as np
import pandas as pd
from google.colab import files

print("=== CALCULADOR DE AUTOVALORES, AUTOVECTORES Y POLINOMIO (20x20) ===")
print("Por favor, sube tu archivo de Excel (.xlsx o .xls)\n")

# 1. Abrir la interfaz nativa de Colab para subir el archivo
archivos_subidos = files.upload()

if archivos_subidos:
    nombre_archivo = list(archivos_subidos.keys())[0]

    try:
        # 2. Leer el archivo Excel en memoria (puros nГєmeros)
        contenido_bytes = archivos_subidos[nombre_archivo]
        df = pd.read_excel(io.BytesIO(contenido_bytes), header=None)
        matriz = df.to_numpy()

        # 3. Validar las dimensiones de la matriz
        filas, columnas = matriz.shape
        if filas != 20 or columnas != 20:
            print(
                f"\nвќЊ ERROR: La matriz subida es de {filas}x{columnas}."
            )
            print("Debe ser estrictamente de 20x20 pГ­xeles.")
        else:
            print(f"\nвњ… Archivo '{nombre_archivo}' cargado correctamente.")
            print("Procesando matriz de 20x20...\n")

            # 4. Calcular Valores y Vectores Propios
            valores_propios, vectores_propios = np.linalg.eig(matriz)

            # 5. Calcular Coeficientes del Polinomio CaracterГ­stico
            # np.poly(matriz) devuelve los coeficientes de det(О»I - A) ordenados desde О»^20 hasta el tГ©rmino independiente
            coeficientes = np.poly(matriz)

            # 6. Mostrar Resultados Formateados
            print("==================================================")
            print("               RESULTADOS DEL ANГЃLISIS            ")
            print("==================================================")

            print("\n--- COEFICIENTES DEL POLINOMIO CARACTERГЌSTICO ---")
            print(
                "P(О») = c_0В·О»^20 + c_1В·О»^19 + ... + c_19В·О» + c_20 = 0"
            )
            print("--------------------------------------------------")
            for i, coef in enumerate(coeficientes):
                grado = 20 - i
                # Evitar mostrar decimales infinitos usando formato cientГ­fico si el nГєmero es muy grande
                if abs(coef) > 1e4 or (abs(coef) < 1e-4 and coef != 0):
                    print(f"c_{i:<2d} (coeficiente de О»^{grado:<2d}) = {coef:.4e}")
                else:
                    print(f"c_{i:<2d} (coeficiente de О»^{grado:<2d}) = {coef:.4f}")

            print("\n--- VALORES PROPIOS (О») ---")
            for i, val in enumerate(valores_propios):
                if np.iscomplexobj(val):
                    print(
                        f"О»_{i+1:02d} = {val.real:.4f} + {val.imag:.4f}j"
                    )
                else:
                    print(f"О»_{i+1:02d} = {val:.4f}")

            print("\n--- VECTORES PROPIOS ASOCIADOS ---")
            print(
                "(Cada columna 'i' corresponde al vector propio de su respectivo О»_i)\n"
            )
            for i in range(20):
                print(f"Vector propio v_{i+1:02d} (asociado a О»_{i+1:02d}):")
                vector = vectores_propios[:, i]
                print(np.round(vector, 4))
                print()

            # 7. ComprobaciГіn Algebraica de la Traza
            traza_matriz = np.trace(matriz)
            suma_lambdas = np.sum(valores_propios)

            print("==================================================")
            print("             COMPROBACIГ“N ALGEBRAICA              ")
            print("==================================================")
            print(f"Traza de la matriz original (ОЈ Diagonal): {traza_matriz:.4f}")
            print(f"Suma de todos los valores propios:       {suma_lambdas.real:.4f}")
            print(
                f"Nota: El coeficiente c_1 debe ser igual a -Traza: {-coeficientes[1]:.4f}"
            )
            print("==================================================")

    except Exception as e:
        print(f"\nвќЊ OcurriГі un error al procesar el archivo: {str(e)}")
else:
    print("\nвќЊ No se seleccionГі ningГєn archivo.")
```
