```python
import io
import cv2
import ipywidgets as widgets
import numpy as np
from IPython.display import display
from PIL import Image

# Crear los componentes de la interfaz
uploader = widgets.FileUpload(
    accept=".jpg,.jpeg,.png",  # Tipos de archivos permitidos
    multiple=False,  # Solo una imagen a la vez
    description="Subir Imagen",
)

output_interface = widgets.Output()


def procesar_imagen(change):
    with output_interface:
        # Limpiar salidas anteriores
        output_interface.clear_output()

        if not uploader.value:
            print("Por favor, selecciona una imagen.")
            return

        print("🔄 Procesando imagen...")

        # Obtener el archivo subido
        uploaded_file = list(uploader.value.values())[0]
        image_content = uploaded_file["content"]

        # Leer la imagen usando PIL y convertirla a RGB
        img = Image.open(io.BytesIO(image_content)).convert("RGB")

        # Mostrar una miniatura de la imagen subida
        img_thumbnail = img.copy()
        img_thumbnail.thumbnail((300, 300))
        print("\n📷 Imagen subida:")
        display(img_thumbnail)

        # Convertir a un array de NumPy
        img_np = np.array(img)

        # Separar los canales R, G, B
        R = img_np[:, :, 0]
        G = img_np[:, :, 1]
        B = img_np[:, :, 2]

        # Calcular la matriz de luminancia (Fórmula estándar ITU-R 601-2)
        matriz_luminancia = (0.299 * R + 0.587 * G + 0.114 * B).astype(np.uint8)

        # --- Resultados ---
        print("\n✅ ¡Matriz calculada con éxito!")
        print(f"Dimensiones de la matriz (Alto x Ancho): {matriz_luminancia.shape}")
        print(f"Valor Mínimo encontrado: {matriz_luminancia.min()}")
        print(f"Valor Máximo encontrado: {matriz_luminancia.max()}")

        # CORREGIDO: 'edgeitems' en lugar de 'edgecolor' para recortar la vista previa
        np.set_printoptions(threshold=100, edgeitems=5, precision=0)
        print("\n📊 Muestra de la matriz de luminancia (filas y columnas extremas):")
        print(matriz_luminancia)

        # Opción para descargar la matriz completa como archivo de texto
        np.savetxt("matriz_luminancia.txt", matriz_luminancia, fmt="%d")
        print(
            "\n💾 Se ha guardado el archivo 'matriz_luminancia.txt' en tus archivos de Colab con la matriz completa."
        )


# Registrar el evento cuando se sube un archivo
uploader.observe(procesar_imagen, names="value")

# Mostrar la interfaz en pantalla
print("╔════════════════════════════════════════╗")
print("║     EXTRACTOR DE MATRIZ DE LUMINANCIA  ║")
print("╚════════════════════════════════════════╝")
display(uploader, output_interface)
```
