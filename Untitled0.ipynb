```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Especial para mostrar imágenes en Colab
from google.colab import files

# ==========================================
# 1. INTERFAZ INTERACTIVA DE CARGA DE IMAGEN
# ==========================================
print("====================================================")
print("  INTERFAZ DE CARGA: SELECCIONA TU RADIOGRAFÍA       ")
print("====================================================\n")

uploaded = files.upload()

# Verificar que se haya subido un archivo
if len(uploaded) == 0:
    print("❌ ERROR: No se seleccionó ningún archivo.")
else:
    # Obtener el nombre del archivo cargado dinámicamente
    filename = list(uploaded.keys())[0]
    file_bytes = np.frombuffer(uploaded[filename], np.uint8)

    # Cargar la imagen en escala de grises para el análisis matricial
    imagen = cv2.imdecode(file_bytes, cv2.IMREAD_GRAYSCALE)

    alto, ancho = imagen.shape
    print(f"\n✅ Imagen '{filename}' cargada con éxito.")
    print(f"Dimensiones de la imagen global: {ancho} x {alto} píxeles.\n")

    # ==========================================
    # 2. ASIGNACIÓN AUTOMÁTICA DE COORDENADAS (SEGURO Y AL AZAR)
    # ==========================================
    # Calculamos una posición proporcional en el cuadrante superior para que caiga en zona pulmonar
    fila_inicio = int(alto * 0.25)
    col_inicio = int(ancho * 0.15)

    fila_fin = fila_inicio + 20
    col_fin = col_inicio + 20

    # ==========================================
    # 3. EXTRACCIÓN DE LA SUBMATRIZ DE ESTUDIO (ROI)
    # ==========================================
    submatriz_A = imagen[fila_inicio:fila_fin, col_inicio:col_fin]

    # ==========================================
    # 4. CONFIGURAR Y MOSTRAR LA MATRIZ NUMÉRICA AUTÉNTICA
    # ==========================================
    print(f"📍 Coordenadas de la ROI: Filas [{fila_inicio}:{fila_fin}] | Columnas [{col_inicio}:{col_fin}]")
    print(f"📐 Dimensiones de la submatriz extraída: {submatriz_A.shape} (Total: 400 píxeles)\n")

    print("=========================================================================")
    print("                      MATRIZ ORIGINAL A (20x20)                          ")
    print("=========================================================================")

    # Formateo para que imprima limpio: columnas alineadas de 3 dígitos
    np.set_printoptions(threshold=np.inf, linewidth=200, formatter={'int': '{:3d}'.format})
    print(submatriz_A)
    print("=========================================================================\n")

    # ==========================================
    # 5. LOCALIZACIÓN VISUAL (DIBUJAR CUADRO ROJO)
    # ==========================================
    # Decodificamos la misma imagen a color (BGR) en memoria para poder pintar en rojo
    imagen_color = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

    # Dibujar el rectángulo rojo (grosor de 2 píxeles)
    cv2.rectangle(imagen_color, (col_inicio, fila_inicio), (col_fin, fila_fin), (0, 0, 255), 2)

    print("📸 Ubicación espacial del recorte dentro de la radiografía (Cuadro Rojo):")
    cv2_imshow(imagen_color)
```
